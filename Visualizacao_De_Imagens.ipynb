```
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
```

1) Mostrar a imagem mostrando a escala da mesma: 

```
# Exibindo imagem
folder = 'lenna.png'

img = cv2.imread(folder, cv2.IMREAD_COLOR)

plt.title('Imagem Original')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

2) Exibindo imagens diferentes tons:

```
img = cv2.imread(folder, cv2.IMREAD_COLOR)
bgr = cv2.split(img)

plt.subplot('221')
plt.title('B')
plt.imshow(bgr[0], 'gray')

plt.subplot('222')
plt.title('G')
plt.imshow(bgr[1], 'gray')

plt.subplot('223')
plt.title('R')
plt.imshow(bgr[2], 'gray')

plt.subplot('224')
plt.title('Original')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

plt.show()
```

3) Visualização dos canais de coloração de forma separada:


```
#%% Load a color image and visualize each channel separately
img = cv2.imread('baboon.jpeg', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

r = [1,0,0] # Vermelho
g = [0,1,0] # Verde
b = [0,0,1] # Azul


c = [0,1,1] # Ciano
m = [1,0,1] # Magenta
y = [1,1,1] # Amarelo

# [ C     [ 1     [ R
#   M   =   1   -   G
#   Y ]     1 ]     B ]

plt.subplot('241');plt.title('RGB');plt.imshow(img)
plt.subplot('242');plt.title('R');plt.imshow(r*img)
plt.subplot('243');plt.title('G');plt.imshow(g*img)
plt.subplot('244');plt.title('B');plt.imshow(b*img)
plt.subplot('246');plt.title('C');plt.imshow(c*img)
plt.subplot('247');plt.title('M');plt.imshow(m*img)
plt.subplot('248');plt.title('Y');plt.imshow(y*img)
plt.show()
```

```
#%% Load a color image and visualize each channel separately
img = cv2.imread('baboon.jpeg', cv2.IMREAD_COLOR)
bgr = cv2.split(img)
colormap = cv2.COLORMAP_JET

plt.subplot('221');plt.title('B');plt.imshow(cv2.applyColorMap(bgr[0], colormap))
plt.subplot('222');plt.title('G');plt.imshow(cv2.applyColorMap(bgr[1], colormap))
plt.subplot('223');plt.title('R');plt.imshow(cv2.applyColorMap(bgr[2], colormap))
plt.subplot('224');plt.title('Original');plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

plt.show()
```

```
# Converting between colors spaces - BGR to YRB - Part I 
# NTSC colorspace - part I

img = cv2.imread(folder, cv2.IMREAD_COLOR)

img2 = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)
yrb = cv2.split(img2)


plt.subplot('231');plt.title('img');plt.imshow(img)
plt.subplot('232');plt.title('img2');plt.imshow(img2)
plt.subplot('234');plt.title('y');plt.imshow(yrb[0])
plt.subplot('235');plt.title('r');plt.imshow(yrb[1])
plt.subplot('236');plt.title('b');plt.imshow(yrb[2])
plt.show()
```

```
#%% Converting between colors spaces - BGR to YRB - Part II 
# NTSC colorspace
img = cv2.imread(folder, cv2.IMREAD_COLOR)

img2 = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
yrb = cv2.split(img2)
diff = cv2.absdiff(gray, yrb[0])
    
plt.subplot('231');plt.title('img');plt.imshow(img)
plt.subplot('232');plt.title('gray');plt.imshow(gray)
plt.subplot('233');plt.title('diff');plt.imshow(diff)
plt.subplot('234');plt.title('y');plt.imshow(yrb[0])
plt.subplot('235');plt.title('r');plt.imshow(yrb[1])
plt.subplot('236');plt.title('b');plt.imshow(yrb[2])
plt.show()
```

```

#%%
# NTSC colorspace
img = cv2.imread(folder, cv2.IMREAD_COLOR)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
#img = cv2.imread('images/rgb_cube2.png', cv2.IMREAD_COLOR)

img2 = cv2.cvtColor(img, cv2.COLOR_BGR2HSV_FULL)
img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
hsv = cv2.split(img2)

plt.subplot('231');plt.title('img');plt.imshow(img_rgb)
plt.subplot('232');plt.title('img2');plt.imshow(img2_rgb)
plt.subplot('234');plt.title('h');plt.imshow(hsv[0], cmap='gray')
plt.subplot('235');plt.title('s');plt.imshow(hsv[1], cmap='gray')
plt.subplot('236');plt.title('v');plt.imshow(cv2.normalize(hsv[2], None, 0, 255, cv2.NORM_MINMAX), cmap='gray')
plt.show()
```
